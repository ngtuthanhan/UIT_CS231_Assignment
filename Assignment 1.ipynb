## Calculate Euclid Distance

```python
from scipy.spatial import distance

def calculate_euclidean_distance(path_img1, path_img2):
    img1 = cv2.imread(path_img1,0)
    size_img1 = img1.shape[0]*img1.shape[1]
    hist_img1 = cv2.calcHist([img1],[0],None,[256],[0,256])
    hist_img1 /= size

    img2 = cv2.imread(path_img2, 0)
    size_img2 = img2.shape[0]*img2.shape[1]
    hist_img2 = cv2.calcHist([img2],[0],None,[256],[0,256])
    hist_img2 /= size
    print("Khoang cach euclidean giua 2 anh {} va {} la: {}".format(path_img1, path_img2, distance.euclidean(hist_img1, hist_img2)))
```

```python
list_img = os.listdir('/content/drive/MyDrive/CS231.M21.KHTN/Baitap/Baitap1/images')
for path_img1_index in range(len(list_img)):
    for path_img2_index in range(path_img1_index+1, len(list_img)):
        calculate_euclidean_distance(list_img[path_img1_index], list_img[path_img2_index])
```

## Calculate Cosine Simarlarity

```python
from scipy.spatial import distance

def calculate_cosine_simarlarity(path_img1, path_img2):
    img1 = cv2.imread(path_img1, 0)
    size_img1 = img1.shape[0]*img1.shape[1]
    hist_img1 = cv2.calcHist([img1],[0],None,[256],[0,256])
    hist_img1 /= size

    img2 = cv2.imread(path_img2, 0)
    size_img2 = img2.shape[0]*img2.shape[1]
    hist_img2 = cv2.calcHist([img2],[0],None,[256],[0,256])
    hist_img2 /= size
    print("Do tuong dong Cosine giua 2 anh {} va {} la: {}".format(path_img1, path_img2, 1- distance.cosine(hist_img1, hist_img2)))
```

## Histogram equalization

```python
def show2image(img_src, img_rst):
    plt.figure(figsize=(20, 20))
    # show img src
    plt.subplot(1,2,1)
    plt.title("Origin Image")
    img_src = cv2.cvtColor(img_src, cv2.COLOR_BGR2RGB)
    plt.imshow(img_src, interpolation='bicubic')

    # show img result
    plt.subplot(1,2,2)
    plt.title("Image After Histogram Equalization")
    img_rst = cv2.cvtColor(img_rst, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rst, interpolation='bicubic')
```

```python
def Equa_hist(source, title = None):
    origin_img = cv2.imread(source)
    b, g, r = cv2.split(origin_img)

    if title is not None:
        clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(title, title))
    else:
        clahe = cv2.createCLAHE(clipLimit=3.0)

    clahe_b = clahe.apply(b)
    clahe_g = clahe.apply(g)
    clahe_r = clahe.apply(r) 

    equa_img = cv2.merge((clahe_b, clahe_g, clahe_r))

    show2image(origin_img, equa_img)
```
